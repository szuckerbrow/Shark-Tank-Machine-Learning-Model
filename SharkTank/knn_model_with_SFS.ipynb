```python
# Select the columns that do not provide the answer to use in the feature selection
sharktank_df_limited = df.loc[:, ['Got Deal','Industry','Pitchers Gender','US Viewership', 'Original Ask Amount',
                                       'Original Offered Equity','Has_Website','Barbara Corcoran Present','Mark Cuban Present',
                                       'Lori Greiner Present','Robert Herjavec Present','Daymond John Present', 'Kevin O Leary Present', 
                                       'Guest Present',]]
```

```python
# Split target column from dataset
y = sharktank_df_limited['Got Deal']
X2 = sharktank_df_limited.drop(columns='Got Deal')
```

```python
X2.head()
```

```python
# Encode the categorical variables using get_dummies
X2 = pd.get_dummies(X2)
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X2, y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Scale the training data
X_train_scaled = StandardScaler().fit_transform(X_train)
X_test_scaled = StandardScaler().fit_transform(X_test)
```

```python
# Instantiate the model
model = KNeighborsClassifier(n_neighbors=40)
```

```python
# Initialize Sequential Feature Selector
sfs = SequentialFeatureSelector(model, n_features_to_select='auto', direction='forward', scoring='accuracy', cv=5)

# Fit SFS to the data
sfs.fit(X_train_scaled, y_train)

# Get selected feature indices
selected_feature_indices = sfs.get_support(indices=True)

# Get selected feature names
selected_feature_names = X2.columns[selected_feature_indices]

# Subset the data with selected features
X_train_selected = X_train_scaled[:, selected_feature_indices]
X_test_selected = X_test_scaled[:, selected_feature_indices]

# Print selected feature indices and names
for index, name in zip(selected_feature_indices, selected_feature_names):
    print(f"Feature {index}: {name}")
```

```python
# Train the model
model.fit(X_train_selected, y_train)
```

```python
# Create predictions
y_pred = model.predict(X_test_selected)

# # Review the predictions
# y_pred
```

```python
# Print confusion matrix
confusion_matrix(y_pred,y_test)
```

```python
# Print classification report
print(classification_report(y_pred,y_test))
```
